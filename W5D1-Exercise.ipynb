## 5.1.1 Data Wrangling and Visualizing

**Exercise 5.1.1.1** Read in the dataset "nchs_death_causes.csv" as a list of lists. You will need to use the `csv` package and then will want to go through each row appending it to a list you create.

```python
import csv

f = open("nchs_death_causes.csv", "r",)
reader = csv.reader(f)
list1=[]
for row in reader:
    list1.append(row)
f.close()

print(list1)
```

**Exercise 5.1.1.2** Make a list that is a subset of the data, containing only one cause of death and one state.

```python
#wyoming = list1[-1][]

temp_list=[]
sub_list=[]

for element in list1:
    if "Wyoming" in element:
        temp_list.append(element)

for element in temp_list:
    if "Kidney disease" in element:
        sub_list.append(element)

print(sub_list)
```

**Exercise 5.1.1.3** Plot a line graph that shows the development of the death rate over time.

```python
import matplotlib.pyplot as plt
values = [int(item[4]) for item in sub_list] #makes a list of values af index 4 in each list inside list
values.reverse() #so it's old -> new
keys = [int(item[0]) for item in sub_list]   #makes a list of keys af index 0 in each list inside list
keys.reverse() # so it's old -> new

WY_kidney = list(zip(keys, values)) #makes a list of tuples based on keys and values

plt.plot(values)
#plt.show()
print(WY_kidney)
```

**Exercise 5.1.1.4** Make your figure self-contained and export it/save it to your machine.

```python
# Making it self-contained
plt.figure(figsize=(10, 7), dpi= 200)

plt.plot(values)

plt.title("Graph to show WY kidney disease death rates by year")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.xticks(range(0,len(keys)), keys, rotation=90)

#plt.show() #OBS: I need to not have show in here, or else the export does not work properly!

#Exporting the graph
import os
file_dir = os.getcwd() #in case we want to save it notebook folder. if not replace with correct link to folder
plt.savefig(file_dir + "\\Kidney_Disease.png") #exports graph and names it
```
