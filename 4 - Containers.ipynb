# Dictionaries

```python
#Declaring a dictionary - remember {}
my_dict={'Samantha':'Anthropology',
         'Friedolin': 'Sociology', 
         'Ulf': 'CS'}
```

```python
my_dict['Friedolin']
```

```python
#Zipping together keys and values (makes a list of tuples)
keys = ['Samantha', 'Friedo', 'Ulf']
ages = [103, 78, 85]
key_values = list(zip(keys, ages))
print(key_values)

#Making the list of tuples a dictionary (note the {})
my_dictionary2 = dict(key_values)
print(my_dictionary2)
```

# Loops

```python
#Example of for loop.
ages = [26, 21, 44, 28, 37]
avg_age=0  #Because I do not yet know the average, i put 0

for i in ages: #The choice of using the letter i is insignificant (could be something else)
    avg_age += i #

avg_age/= len(ages) #len = length (here of ages), to find the average
print(avg_age)
```

```python
# On operations used above (the +=). This below gives the value -6
my_var = 0
my_var = my_var - 2
my_var *= 3

print(my_var)
```

```python
#On variables not being able to have the same name. Therefore, the below variable is 1, because it overwrites.
my_var2 = 5
my_var2 += 3.8
my_var2 = 1

print(my_var2)
print(type(my_var2))
```

```python
#Example of while loop
count = 0
avg_age = 0

while count<len(ages):
    avg_age += ages[count]
    count += 1
avg_age /= len(ages)
print(avg_age)
```
