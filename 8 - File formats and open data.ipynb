# Data formatting

Data is versatile and can be represented in different ways.

Three main classes of data storage:
- Tabular (CSV)
- Non-tabular (JSON, XML)
- Associative (networks)
    - Where data is best stored as a network. E.g. Facebook store their user data as a big network.

## Tabular data

Common formats:
- CSV, TSV, SCSV, SSV (remark that they all end in SV - which stands for 'separated values')
    - CSV is super common. Because if everybody uses CSV, nobody accidentally uses commas in their data (it would mess up the tabular format of the data)
        - Therefore, use a . if intention is not to separate data (e.g. in big numbers)

Different delimiters can be used
- Often commas are used, but also tabs and semicolons

Missing values are OK
- Can be solves with just adding two commas, if there is no data in a specific tab (e.g.: , , )

Exists because: data has been processed into this format, or database assumed no change in formatting
- Often times 'raw' data are not stored in this format, e.g. in businesses. This is because tables are an inflexible format, where it is difficult to add new features along the way.

Tabular data is the backbone of SQL-type databases (powering most data storage for the web)
- SQL databases interlink tables of data, each one a file that can be retrieved in e.g. CSV format.
    - SQL means data is stored as tables.

## Non-tabular data

Common formats: 
- JSON, XML
    - XML is a precursor to HTML, but is not really used anymore. JSON is more eady to read as well.

Tree-like data structure
- 'Nested lists and dictionaries'
- Data points can have different and variable number of attributes (therefore, more flexible than tables)
    - Due to this, inconsitency and missing values are expected.
- Tree structure means that two nodes from different trees cannot be connected
    
Exists because: data is relatively raw from source, and databse assumed formatting would change with time

# Open data
Gist:
- People, organizations and nations share data for various reasons
- Two main ways to offer up data for open use:
    - Through an API
        - Can be thought of as a waiter at the restaurant - the front-end that presents products from the back-end
    - As a downloadable file

#### Open data - places on the web we can get it
- Geographical data
    - OpenStreetMap (has downloadable datasets in tabular formats)
- City data
    - Chicago Data Portal (used for criminological research)
    - City of Copenhagen (e.g. positions of streetlamps - but not crime data)
- Political data
    - Congress.gov (votes, bill status)
    - Oda.ft.dk (Folketinget API for accessing e.g. voting records)
- Research data (offered by researchers)
    - SocioPatterns (e.g. samples co-presence among French people)
    - Supreme Court Citation Network Data (references to past cases that make precedence for a given case - network analysis)
    - Enron Email Dataset (all internal emails from Enron organization)
    - Coronaviridae (more biotechnical, but data on experiments being done)
- Competition datasets
    - Kaggle (competition about solving a company's problem)
- Transactional data (monetary)
    - Block Explorer (blockchain technology made transactions public - but no knowledge of who made the transaction)
    - Etherscan
