## 6.1.1 Data Wrangling and Visualizing

**Exercise 6.1.1.1** Read in the dataset "nchs_death_causes.csv" using `pandas` methods.

```python
import pandas as pd 
us_death = pd.read_csv("nchs_death_causes.csv")
print(us_death)
```

**Exercise 6.1.1.2** Make a `DataFrame` that is a subset of the data, containing only one cause of death and one state.

```python
unintentional = us_death[us_death["Cause Name"].isin(['Unintentional injuries'])]
wisconsin = unintentional[unintentional["State"].isin(['Wisconsin'])]

wisconsin #Now we have only unintentional injuries for wisconsin in this dataframe
```

**Exercise 6.1.1.3** Plot a line graph that shows the development of the death rate over time. Pay attention to whether your plot is intuitive.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(wisconsin['Age-adjusted Death Rate'][wisconsin['Cause Name'] == 'Unintentional injuries'], 'o')                                           
```

**Exercise 6.1.1.4** Make your figure self-contained and export it/save it to your machine.

```python
years = [int(item) for item in wisconsin['Year']] 
years.reverse()

deaths = [item for item in wisconsin['Age-adjusted Death Rate']]
deaths.reverse()
```

```python
plt.figure(figsize=(10, 7), dpi= 200)

plt.plot(years, deaths, 'o')

plt.title("Graph to show WI unintentional injury deaths")
plt.xlabel("Year")
plt.ylabel("Age-adjusted death rate")
plt.xticks(years[0:18], rotation=90)

#Export
import os
file_dir = os.getcwd() #in case we want to save it notebook folder. if not replace with correct link to folder
plt.savefig(file_dir + "\\Unintentional Injury WI.png")
```

**Exercise 6.1.1.7** Make a `DataFrame` that is a subset of the data, containing data for only one year and one state.

```python
#texas_2016 = [us_death[us_death["State"]=["Texas"] and [us_death[us_death["Year"]=[2016]]]]

colorado = us_death[us_death["State"].isin(['Colorado'])]
colorado_2016 = colorado[colorado["Year"].isin([2016])]

colorado_2016 = colorado_2016.drop([57], axis = 0)

colorado_2016
```

**Exercise 6.1.1.8** Create a  _self-contained_ bar graph showing the death rate across causes of death in your subset of the data.

```python
deathrates2016 = [item for item in colorado_2016["Cause Name"]]
causes2016 = [item for item in colorado_2016 ["Age-adjusted Death Rate"]]

plt.figure(figsize=(10, 7), dpi= 120)
plt.bar(deathrates2016, causes2016)

plt.title("Bar chart showing death rate per cause for 2016, CO")
plt.xlabel("Cause")
plt.ylabel("Deaths")
plt.xticks(rotation=90)
```
