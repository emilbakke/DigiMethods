# Application Programming Interfaces (APIs)

What are they?
- Interface to exted functionality
- Mechanisms to regulate data access (controls the data one might get access to - good for private companies)
- Pipelines for data access
- Tools for data processing

### How do I use them?
- Manually:
   - Generate API calls (see slide for picture of API call URL)
   - Parse returned data
- Use available tools/libraries

#### Seeking out APIs
Question 1: Does one exist?
- DBA.dk (no API)
- Boligportal (no API)
- MobilePay (API)

Question 2: What functionality does it offer?

Question 3: Is there a module to use it?


Some API examples
- Google Vision API (identify objects in images. Good for multiple photographs. But you would probably have to pay for access to the API)
- ABBYY OCR (OCR-ing a scanned PDF)

#### Important concepts
- Authentification/credentials
- Rate limiting
    - Twitter is relatively open API - but it can change into the likes of Facebook, etc.
- Algorithmic drift
    - APIs change. So applying a script written two years ago, that worked then, may not work now.



# Ethics

Detailed discussion of ethics in "Bit by Bit" by Matthew Salganik
- Freely available online
- Widely used in computational social science in regards to ethics
- Watch lecture on Youtube Salganik did on ethics

Why care about ethics?
- Fear
    - Research not getting me into trouble
    - How research will be looked upon by society and peers
    - Field not getting into trouble because of research (relevant in HK case)
- Hope
- Necessity
    - Bound to be ethical questions about using data like this
        - E.g. people not 'knowing' they participate, or do so 'willingly'

### Three approaches

Rules-based approach
- Go with law and regulations as ethical guidelines

Ad hoc approach
- Often comes from fields not focusing on human subjects, e.g. computer science
    - Coming up with stuff on the fly

Principles-based approach
- Outline principles in approaching research, and as a way to communicate about ethics

### Prominent examples
Emotional contagion
- Bocking out certain content from people's Facebook timeline
    - Tendency to block out positive or negative content
    
Tastes, Ties, and Time
- Collected information on soon-to-be-students on Facebook, and then took out names and addresses and released the data to the public. However, it was still possible to point out the student that come entry had come from.

Encore
- Studied censorship around the world. Wrote a cookie integrated into websites. Cookie would then tell researchers when a user was censored on e.g. Wikipedia
- No information on human individuals. So review board did not want to review it. Problem is that the information may still have implications for human individuals.

### Principles
Respect for persons
- E.g. researcher is not the one to decide who wants to participate. Think of informed consent.

Beneficence (godg√∏renhed)
- Minimize risks, maximize benefits

Justice
- Distribute burdens and benefits as equally as possible

Respect for Law and Public Interest
- Can be complicated with international data gathering - which country's Law applies when I collect American data?
- Some researchers argue that the public interest of their research were bigger than the problems associated with breaking e.g. Google's terms of conditions (HK and Twitter here?)
    - Remember, these are principles and can be bent and argued around differently
