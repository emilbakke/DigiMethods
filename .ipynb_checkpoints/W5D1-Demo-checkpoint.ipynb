# Basic visualizations
(Potential) components:
- Axes (e.g. X and Y - sometimes Z)
- Ticks
    - Marks that define values on axis
- Labels
    - Labels for the ticks. E.g. 'Low' for a tick on an X axis
- Titles
    - Often there are not titles in journal article visualizations, instead they describe it
- Colors
- ...
    - Multiple occurences at a particular point can be shown in different ways. E.g. changing size, or opaqueness
- Facets
- Legend
    - E.g. what color 1 means, color 2 means, etc.

### Line plots
Probably most common visualization type
- Observations are linked by a trendline
    - Good for observations over time
    - Can sometimes be a good idea to add dots for each individual observation
- Always remember adding the origin (lowest number on Y axis) to increase readability/reduce misunderstandings

### Bar Graphs
Specific observations and a value represented for each observation by height of bar
- Remember labels on Y-axis to further understanding of bar graphs
- Again, remember the origin (often they are 0 in these types of graphs)

Useful for data that is not trying to show development over time, e.g. 'snapshots' of time

### Scatter plots
Specific observations and visualizing two different variables for an observation
- E.g. for each US state (the observation) the variables could be population and crime rate
- Often diffrent observations will have the exact same value. Then one should think about how to distinguish these by size, opaqueness, etc.

### Histograms
Also bars, but not showing specific values for an observation. Instead showing for all the observations, how many take on a specific value.
- Gives a sense of the distribution of a variable
- The pin size needs to be considered. If pin is small it will give a lot of bars and be harder to interpret. But with big bars, you  get bars that span across a bigger interval than the smaller alternative, thus potential to misinterpret precisiveness.

## Best Practices
Ink-to-Information Ratio
- E.g. when people use 3D models for bar plots, it adss more ink but not more information (maybe even less)
- In short, do not add anything that is not necessary

Self-contained
- Should be able to comprehend on its own (out of context)
    - E.g. by adding titles, labels, ticks, etc.

Accessible
- Make sure everybody can read it. 
    - E.g. take into account colorblindness (Colorbrewer website can help find color palettes)

Intuitive
- In most cases: go with the way you would intuitively read the figure
    - Often this can be based on existing conventions within visualizations. E.g. old -> new on x-axis

## 5.1.2 Formatting Graphs

Adding a title, axis, and tick labels for comprehension

```python
values = [item[1] for item in flu_death_rates]
labels = [item[0] for item in flu_death_rates]

plt.bar(labels, values)

plt.title("Bar chart to show US 2016 flu/pneumonia death rates by state")
plt.xlabel("States")
plt.ylabel("Deaths per 100,000")
plt.xticks(rotation=90) #rotates x-axis labels, so they dont overlap

plt.tight_layout()
```

Adjusting the plot size

```python
values = [item[1] for item in flu_death_rates]
labels = [item[0] for item in flu_death_rates]

plt.figure(figsize=(10, 7), dpi= 120) #tells it to set up a figure in a certain size. Dpi gives higher resolution

plt.bar(labels, values)

plt.title("Bar chart to show US 2016 flu/pneumonia death rates by state")
plt.xlabel("States")
plt.ylabel("Deaths per 100,000")
plt.xticks(rotation=90)

plt.tight_layout()
```

## 5.1.3 Exporting / Saving Graphs

A self-contained line plot.

```python
values = [item[1] for item in NY_flu_death_rates]
labels = [item[0] for item in NY_flu_death_rates]

plt.figure(figsize=(10, 7), dpi= 120)

plt.plot(values)

plt.title("Graph to show NY flu/pneumonia death rates by year")
plt.xlabel("Year")
plt.ylabel("Deaths per 100,000")
plt.xticks(range(0,len(labels)), labels, rotation=90)

plt.show()
```

Exporting the line graph.

```python
import os

values = [item[1] for item in NY_flu_death_rates]
labels = [item[0] for item in NY_flu_death_rates]

plt.figure(figsize=(10, 7), dpi= 250)

plt.plot(values)

plt.title("Graph to show NY flu/pneumonia death rates by year")
plt.xlabel("Year")
plt.ylabel("Deaths per 100,000")
plt.xticks(range(0,len(labels)), labels, rotation=90)

file_dir = os.getcwd() #in case we want to save it notebook folder. if not replace with correct link to folder

if '/' in file_dir:
    print("I am a Mac.")
    plt.savefig(file_dir + "/Flu_Deaths_NY_2016.png") 
else:
    print("I am a Windows PC.")
    plt.savefig(file_dir + "\\Flu_Deaths_NY_2016.png") #this is where name should be specified, and export happens.
```
